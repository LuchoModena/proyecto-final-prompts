# Sistema Inteligente de Planificación Académica basado en Prompt Engineering

El presente proyecto tiene como objetivo desarrollar un sistema inteligente de planificación académica basado en técnicas de Prompt Engineering. La problemática abordada es la desorganización y baja productividad en estudiantes universitarios, una situación frecuente que impacta negativamente en el rendimiento académico y el bienestar emocional.

Para resolver este problema, se propone la implementación de un sistema de prompts que permita analizar el contexto personal del estudiante, detectar sobrecargas, priorizar tareas y generar planes de estudio personalizados. Además, se incorpora la generación de imágenes motivacionales mediante modelos texto-imagen, con el fin de reforzar la adherencia al plan. El proyecto se implementa como una prueba de concepto (POC) en un entorno de Jupyter Notebook.

Problema a abordar: Muchos estudiantes universitarios presentan dificultades para organizar su tiempo de estudio de manera eficiente. Esta desorganización suele estar asociada a factores como la carga laboral, múltiples materias, falta de planificación estratégica y desconocimiento de técnicas de estudio.

Como consecuencia, se generan situaciones de estrés, bajo rendimiento académico y dificultad para cumplir objetivos educativos. Dado que esta problemática es frecuente y transversal a múltiples disciplinas, resulta relevante desarrollar herramientas que faciliten la planificación académica de manera personalizada.

Propuesta de solución: La solución propuesta consiste en el desarrollo de un sistema basado en inteligencia artificial que, mediante el uso de prompts diseñados estratégicamente, permita: Analizar el contexto del estudiante - Detectar sobrecarga y puntos críticos - Priorizar materias según dificultad y fechas de examen - Generar un plan semanal optimizado - Ajustar dinámicamente el plan en función de nueva información

Adicionalmente, se incorporan prompts de generación de imágenes con el objetivo de crear recursos visuales motivacionales que acompañen el proceso de estudio.

Objetivos específicos: Diseñar prompts optimizados para el análisis de contexto académico - Implementar un sistema de generación de planes de estudio personalizados - Incorporar criterios de priorización y gestión de carga cognitiva - Generar recursos visuales motivacionales mediante IA - Evaluar la efectividad del sistema mediante distintos casos de uso

- Metodología: El proyecto se desarrolla como una prueba de concepto (POC) basada en un flujo estructurado de interacción con modelos de inteligencia artificial.

- Flujo del sistema
Recolección de datos --> Se obtiene información relevante del estudiante:
Materias - Horarios disponibles - Carga laboral - Objetivos académicos - Fechas de examen - Nivel de dificultad percibida - Validación del contexto
El sistema solicita información adicional en caso de ser necesario, evitando respuestas incompletas.

Análisis de la situación
Se evalúa: Nivel de carga semanal - Riesgo de sobrecarga - Prioridad de materias - Generación del plan
Se construye un cronograma semanal que incluye:
Bloques de estudio profundo - Espacios de repaso - Momentos de descanso - Ajuste dinámico
El sistema permite iterar sobre el plan en función de nuevas condiciones o feedback del usuario.

Herramientas y Tecnologías:
Modelos de lenguaje (ChatGPT) → texto-texto
Generación de imágenes (NightCafe) → texto-imagen
Jupyter Notebook → implementación de la POC
Python → estructuración del flujo de prompts

Técnicas de Prompt Engineering utilizadas: Definición de rol (role prompting) - Contextualización detallada - Descomposición de tareas - Especificación de formato de salida - Validación previa de información - Incorporación de criterios de decisión (prioridad, energía, dificultad)

```python
def generar_prompt(contexto):
    prompt = f"""
Actúa como un coach profesional experto en productividad académica.

Contexto del estudiante:
{contexto}

Tareas:
1. Analiza la carga semanal considerando horas disponibles.
2. Detecta puntos de estrés o sobrecarga.
3. Prioriza materias según dificultad y cercanía de exámenes.
4. Considera niveles de energía (alta, media, baja) según días.
5. Diseña un plan semanal equilibrado.
6. Incluye bloques de estudio profundo, repaso y descanso.
7. Justifica las decisiones estratégicas.

Si falta información, solicitála antes de generar el plan.

Entrega en formato:
- Explicación en texto
- Tabla organizada por día y horario
"""
    return prompt
```
